```
# Append data

import pandas as pd, numpy as np

# Concatenate series
ser1 = pd.Series(['A','B','C'], index=[1,2,3])
ser2 = pd.Series(['D','E','F'], index=[4,5,6])
ser_concat=pd.concat([ser1,ser2])
print('Concatenate Series:\n', ser_concat)
print("\n")


# Concatenate dataframes
df1 = pd.DataFrame({'B1': [1, 2, 3], 'B2': [2, 5, 8], 'B3': [3, 6, 9]})
df2 = pd.DataFrame({'B1': [4, 5, 6], 'B2': [10, 9, 8], 'B3': [4, 8, 5]})
df_concat=pd.concat([df1, df2]).reset_index(drop=True)
print('Concatenate DataFrames:\n', df_concat)
print("\n")

# With Append() method
ser_concat1 = ser1.append(ser2)
df_concat1 = df1.append(df2 , ignore_index=True)
print('Append Series:\n', ser_concat1)
print("\n")
print('Append DataFrames:\n', df_concat1)
print("\n")

df_concat2 = df1.append({'B1': 4, 'B2': 10, 'B3': 4} , ignore_index=True)
print('df_concat2:\n', df_concat2)
print("\n")

# Adding lines with a loop
df3 = pd.DataFrame({'B1': ["A1", "A2", "A3"], 'B2': [2, 5, 8], 'B3': [3, 6, 9]})
print(df3)

for i in range(4,6):
     df3.loc[i] = ['A' + str(i)] + [i+1, i+2]

print(df3)
```
