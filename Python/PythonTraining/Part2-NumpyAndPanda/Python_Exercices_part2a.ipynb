```python
# Which employee set the record for the highest number of sales in one month this year? What is this record?
result1 = df.sort_values(by='nVentes', ascending=False)
#print(result1)
print(type(result1)) # <class 'pandas.core.frame.DataFrame'>
print(type(result1.head(1).EmpID)) # <class 'pandas.core.series.Series'>
print(type(result1.head(1).EmpID.values)) # <class 'numpy.ndarray'>
print(result1.head(1).EmpID.values[0])
print(type(result1.EmpID)) #<class 'pandas.core.series.Series'>
print(result1.EmpID.values[0])
print(type(result1.iloc[0]))
print(type(result1.iloc[0][0]))
print(result1.iloc[0][0])
print(type(result1.iloc[0][1]))
print(result1.iloc[0][1])
print(result1.head(1))

```

```python
# Calculate the average amount of sales for each employee each month.
df['avgVente'] = df.TotVentes / df.nVentes
df
```

```python
# Calculate the average amount of each employee's sales for the entire year.
result2 = df.groupby('EmpID')['nVentes', 'TotVentes'].sum()
type(result2)
print(result2)
result2['avgAmout'] = result2.TotVentes / result2.nVentes
print(result2)
```
