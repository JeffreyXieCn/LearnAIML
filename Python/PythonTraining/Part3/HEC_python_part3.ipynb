# Packages for data visualization

## Matplotlib

**Matplotlib** is a complete library for the creation of ** static, animated and interactive Python** visualizations. It is possible to :

   - Create, make graphics in a few lines of code
    
   - Customize, take full control of a graph's attributes 
    
   - Sharing, making our visualizations accessible to everyone



## Seaborn

**Seaborn** is a **data visualization library** in Python based on matplotlib. It provides a high level interface for drawing attractive and informative statistical **graphs**. Here are some of the features Seaborn offers:

   - A dataset-oriented API for examining relationships between multiple variables

   - Specialized support for the use of categorical variables to present observations or aggregate statistics

   - Options to view single and double variable distributions and compare them between subsets of data

   - Estimation and automatic plotting of linear regression models for different types of dependent variables

   - Practical views on the overall structure of complex data sets

   - High-level abstractions for structuring multi-plot grids that make it easy to build complex visualizations

   - Concise style control of matplotlib miniatures with several integrated themes

   - Tools to choose color palettes that accurately reveal the patterns in your data

##Comparison of Matplotlib and Seaborn

**Matplotlib** : Matplotlib is the basis for static plotting in Python. Many other visualization tools are built on top of it, such as seaborn and Pandas DataFrames plot method. It is versatile meaning it is able to plot anything, but non-basic plots can be very verbose and complex to implement. On the other hand, basic plots such as histograms and scatter plots are very easy to do. 

If your goal is to make simple and quick basic plots, there is no big drawback. However, for more complex such as pair plots and heat maps, it is interesting to use some higher-level tools.

The matplotlib’s website contains a very complete documentation, some tutorials and a huge variety of examples, which makes its use a lot easier.

**Seaborn** : Seaborn is the go-to tool for static plotting. It integrates very well with Pandas DataFrames, making it possible to assign column names to the axis, which makes the code clearer. The plots are naturally prettier and easy to customize with color palettes. There are many built-in complex plots like cluster maps, which are very convenient for analyzing data quickly and effectively.

Making complex plots with low effort is accomplished by some features like automatic labels for the axes and grouping with specialized support for categorical variables. Even complex tasks like multi-plotting are abstracted to a high-level grid structure.

Documentation is probably everybody's favorite thing about Seaborn. Everything is minimalist and very well organized. All parameters are well explained in a simple language and didactic examples are provided along with the documentation page, demonstrating how some parameters can affect the visualization. So, everything you need is in one place, and in a non-overwhelming way.

## Dot cloud ##

A scatterplot is a graph between two **continuous variables**. Each observation is represented by a point. Several options/arguments are available to customize our graphs. Here we use the function `scatterplot` from the seaborn library. 

Here are different visualizations by adding some arguments : 

- `hue`. Takes as input the **name of a column of the DataFrame**. Allows to produce displays (lines, points, boxes...) of **different colors** from **numeric** or **categorical** variables according to their **grouping**. 


- The `xscale` argument is used to activate a logarithmic scale on the abscissa axis. This scale is particularly adapted to account for orders of magnitude in applications. It shows in a small space a wide range of values, provided they are non-zero and of the same sign.

# Matplotlib example

Here a version of a dot cloud graph using Matplotlib. You can easily compare the code used here and the one used to produce a similar plot with Seaborn. Adding the colors for each continent in Matplotlib would necessitate a For loop and a dictionary to create the same effect as the simple Seaborn code. 

```
# Simple dot cloud with a predefined style using Matplotlib
# create a figure and axis
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(df_1992['gdpPercap'], df_1992['lifeExp'])
# set labels and log scale for X axis
ax.set_xlabel('gdpPerCap')
ax.set_ylabel('lifeExp')
ax.set_xscale('log')
ax.set_title("Dot cloud grap with Matplotlib")
colors = {'Asia':'r', 'Oceania':'g', 'Africa':'b'}
```

```
# Simple point cloud with a predefined style using SeaBorn
sns.set_style("darkgrid")
ax = sns.scatterplot(x = "gdpPercap", y = "lifeExp", data = df_1992, hue = "continent")
ax.set(xscale = 'log')
# Our 1st continuous variable is GDP per capita on the x-axis.
# The second is the average life expectancy on the ordained axis. 
```

Let's see together other functions/arguments to customize your graph: 

- `set`. Allows you to display a multitude of details on the graph. We will mainly use `title` to add a title, `xlabel` to name the x-axis, `ylabel` for the y-axis. See the `Matplotlib' documentation for additional arguments. 


- `set_xlim` and `set_ylim`. Allows us to create an orthonormal marker of our choice by specifying the lower and upper bound. 


- `axhline` and `axvline`. Allows us to add an x or y constant to our graph. It is possible to specify the `label` or the style with `linestyle`. 

```
ax = sns.scatterplot(x = "gdpPercap", y = "lifeExp", data = df_1992, hue = "continent")
# "hue" is an argument that adds color according to the filters in the column. 
ax.set_xlim(0,28_000)
# Changing the abscissa axis to test the best possible display. 
# Adding a title and a legend
ax.set(title = "Change in GDP per capita as a function of life expectancy in 1992", \
       ylabel = "Life expectancy", xlabel = "GDP/Capita")
# It is possible to add a horizontal or vertical constant line.
# We will first look at a horizontal constant.
ax.axhline(y = 65, label = "World average life expectancy", linestyle = '--')
ax.legend()
# The "legend()" method displays the legend.
# ax.axvline gives a constant vertical line
```

## Graph in curves

A line graph is a **set of points** (visible or not) on a **X-Y grid**, usually connected by a **line segment**. 

They are useful for **representing data over a period of time**: changes in sales, population, inflation, employment, tourism, the price of a commodity, exchange rates, stock market prices, website audience, etc. This type of graph is often **combined** with another type of graph (e.g. bar graph, area graph).

The `Seaborn` library offers the `lineplot()` function to create line graphs. Let's see an example with different arguments : 

- `style`. A grouping variable that will produce lines with different dashes and/or markers. Can have a numeric type but will always be treated as a category.


- `err_style`. Allows confidence intervals to be drawn with translucent error bands or discrete error bars.


- `hue_order`. Ability to give a caption order to be displayed for items related to `hue`, i.e. hue. 



```
ax1 = sns.lineplot(x = "year", y = "gdpPercap", data = df_g, hue = "continent", \
                   style = "continent", err_style = "band", \
                   hue_order = ["Oceania","Europe", "Asia","Americas","Africa"])
```

## Bar graph 

A **bar chart** is a graph that presents **categorical variables** with **rectangular bars** with heights or lengths **proportional** to the values they represent. The bars may be drawn **vertically or horizontally**.

A bar chart shows comparisons between **discrete categories**. One axis of the diagram shows the specific categories being compared and the other axis represents a measured value. Some bar graphs have **clustered bars**, showing the values of **several measured variables**.

The `Seaborn` library provides `barplot` to create our bar graphs. Here is an example with some arguments: 

- "order". **Order the display** of the bar charts. 
- The "orient" argument. Receives "h" or "v" as input. Determines the **orientation** of our bar charts. "v" is selected by default. If "h" is selected, the x axis must be swapped with the y axis. 


```
ax2 = sns.barplot(x = "continent", y = "gdpPercap", data = df_1992, \
                 order = ["Oceania","Europe", "Asia","Americas","Africa"],\
                 orient = "v")
# For the year 1992, Oceania is the continent with the highest GDP per capita in the world.
```

## Histograms

In statistics, a **histogram** is a graphical representation that illustrates the distribution of a **continuous variable** by representing it with **vertical columns**.

`Seaborn` provides us with the `distplot` function to make histograms. Let's see an example with some different arguments:

- `a`. Receives as input the observed data. If it is an object in the series with a name attribute, the name will be used to label the data axis. According to Seaborn, the function would suggest that we receive input from a series in `np.array()` format for better display stability. 


- `bins`. Input an integer. Specifies the number of rectangles to display in our histogram. 


- `hist`. Receives input as a boolean. Tells the function whether to display histograms.


- `rug`. Receives input as a boolean. Tells the function whether to display a `rugplot' on the x-axis. A `rugplot' is a plot of data for a single quantitative variable, displayed as marks along an axis. It is used to visualize the distribution of the data. As such, it is analogous to a histogram with zero-width boxes, or a one-dimensional scatterplot.


- It is called `axlabel'. Receives as input a string of characters. Displays the name of the x-axis. 


- `color`. Receives input from the `Matplotlib' library. 

```
import numpy as np

ax3 = sns.distplot(a = np.array(df_1992["lifeExp"]), bins = 20, hist = True, rug = True, \
                  axlabel = "Life Expectancy", color = "royalblue")
# We note that the distribution is not standardized between the different countries. 
```

## Boxplot ##

A **boxplot** shows the distribution of quantitative data in a way that facilitates comparisons between variables or between levels of a categorical variable. It is then possible to visualize several distribution parameters of a variable: the **median**, **interquartile range** and the **maximum** and **minimum** value of the distribution.

Still in the same library, we will use `boxplot` to display our moustache diagrams. Here is a non-exhaustive list of arguments for this function : 

- `saturation`. Takes as input a real. Determines the **proportion** of the original saturation to draw the colors. Blob`s are often better looking with slightly desaturated colors, but set this value to 1 if you want the plot colors to match the input color specification perfectly.


- `linewidth`. Takes a real input. Determines the **width** of the gray lines that frame the elements of the ``boxplot''.


- showmeans. Takes a boolean as input. Displays the **average** for each ``boxplot''.

Legend of a boxplot graphic :

```
ax4 = sns.boxplot(x = "continent", y = "gdpPercap", data = df_1992, \
                 saturation = 0.8, linewidth = 1, showmeans = True)
# We'll see later how to use filters
```

## Export a graph

After an analysis in Python, it is often necessary to **export our graph**. 
For this, the `savefig` function of the `Matplotlib` library allows us to export a graph easily with several useful arguments. However, and with the latest versions of Seaborn, we should not forget to get the figure with the `get_figure()` function. 




Here are some useful arguments, including `fname` which is indispensable: 

- `fname`. Takes as input a string, a path or the name followed by the format of the document you want to export.


- `facecolor`. Takes as input the colors of the `Matplotlib` library. Allows you to color the external part of the figure. 


- It can also be used to set the `quality'. Takes as input a real between 0 and 95. Determines the **image quality**, on a scale of 1 (worst) to 95 (best). Applicable only if the format is jpg or jpeg, otherwise ignored. 

```
fig4 = ax4.get_figure()
fig4.savefig("boxplot.jpeg", dpi = None, facecolor = "white", quality = 95)
```

## Interactive visualization with Bokeh

What is Bokeh? Bokeh is an interactive visualization library for modern web browsers. It provides elegant, concise construction of versatile graphics, and affords high-performance interactivity over large or streaming datasets. Bokeh can help anyone who would like to quickly and easily make interactive plots, dashboards, and data applications. There are built-in tools that can be included on a widget box attached to the plot and used to explore the data in an interactive way, such as zooming in, selecting and overlaying a crosshair. Bokeh supports unique visualizations like Geospatial plots, Network graphs, etc. right out of the box. If you want to show these visualizations in a browser, there are options available to export them and you can also use it through JavaScript itself!

```
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Call once to configure Bokeh to display plots inline in the notebook.
output_notebook()
```

```
N = 4000
x = np.random.random(size=N) * 100
y = np.random.random(size=N) * 100
radii = np.random.random(size=N) * 1.5
colors = ["#%02x%02x%02x" % (r, g, 150) for r, g in zip(np.floor(50+2*x).astype(int), np.floor(30+2*y).astype(int))]

p = figure()
p.circle(x, y, radius=radii, fill_color=colors, fill_alpha=0.6, line_color=None)
show(p)
```
